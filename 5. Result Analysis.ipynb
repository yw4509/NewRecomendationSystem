```
path="drive/My Drive/1001_project_Total/Data/"
df=pd.read_csv(path+"20170101_sknn.csv")
articles=df["id"].unique()
d={}
for i in range(len(articles)):
    article=articles[i]
    d[article]=i
df["articleId"]=df["id"].map(d)
```

# Category

```
path="drive/My Drive/1001_project_total/wrong/category/"
```

## ItemKNN

```
itemknn=pd.read_csv(path+"itemknn_wrong.txt",sep=",",names=["sessionNo","rank"])
itemknn["sessionNo"]=itemknn["sessionNo"].apply(lambda x:int(x[1:]))
itemknn["rank"]=itemknn["rank"].apply(lambda x:x[:-1])
selected=pd.merge(df,itemknn,on="sessionNo",how="inner")
```

```
a=selected.groupby("deviceType").size()
b=df.groupby("deviceType").size()
device=pd.concat((a,b),axis=1)
device["prop"]=device[0]/device[1]
print("The type of device with ascending number of wrong recommendation: ")
device.sort_values("prop")
```

```
import math
a=selected.groupby("category_num").size()
b=df.groupby("category_num").size()
device=pd.concat((a,b),axis=1)
device["prop"]=device[0]/device[1]
print("The type of device with most number of wrong recommendation: ")
device.sort_values("prop",inplace=True)
k=device[["prop"]].index.values
for i in range(len(k)):
  if i<=10:
    print("The category number:")
    print(k[i])
    print("The category name:")
    print(selected[selected["category_num"]==k[i]]["category1"].unique())
```
